# Notebook to plot data from netCDF file

```python
import xarray as xr
import hvplot.xarray
```

```python
example_run = xr.open_dataset('/Users/vonw/work/software/CommunityFirnModel/CFM_main/CFMoutput_example/csv/CFMresults.nc')
example_run
```

## Plot temperature as a function of time and depth

```python
# ....Toss first value because it has a strange date; restrict to 1980 to 2020
example_run.sel(time=slice(1980, 2020)).hvplot(x='time', y='depth', z='temperature', clim=(220,260))
```

## Plot d18O as a function of time and depth

```python
example_run.sel(time=slice(1980, 2020)).hvplot(x='time', y='depth', z='isotopes_d18O', clim=(-50,0))
```
